#### Heatmap for plotting the UC-CD enrichment analysis results.

```python
import anndata
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns; sns.set_theme()
```

```python
df_CD = pd.read_table("/mnt/2021Jan02/figures/forest_GCST004132_CD.tsv")
df_UC = pd.read_table("/mnt/2021Jan02/figures/forest_GCST004133_UC.tsv")
```

```python
#new_CD = df_CD[df_CD["log OR"]>=-1]
#new_UC = df_UC[df_UC["log OR"]>=-1]
```

```python
new_comb = pd.merge(df_CD,df_UC, left_index=True, right_index=True)

new_comb = new_comb.loc[new_comb["log OR_x"]>0]
new_comb = new_comb.loc[new_comb["FDR_x"]<0.1]
new_comb = new_comb.drop([ 'CI (lower 95%)_x', 'CI (upper 95%)_x', 'P-value_x',
      'CI (lower 95%)_y', 'CI (upper 95%)_y',"FDR_x","FDR_y",
       'P-value_y'], axis=1)
new = new_comb.sort_values(["log OR_x", "log OR_y"], ascending=[False, False])
new= new.drop([ 'LYVE1+ Macrophage',], axis=0)
new.columns = ['CD', 'UC']
```

```python
new
```

```python
import seaborn as sns

sns.set(rc={'figure.figsize':(1.2,12)})
ax = sns.heatmap(new,linewidths=.5, cmap="inferno")
ax.figure.savefig( "/mnt/2021Jan02/figures/FDR.pdf", bbox_inches="tight")
```
